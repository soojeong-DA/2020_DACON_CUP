## train set

```python
train1 = pd.read_csv('data/1차_open_data/train.csv', encoding='cp949')
train1
```

```python
train2 = pd.read_csv('data/2차_open_data/2차_train.csv', encoding='cp949')
train2
```

### 일 단위로 병합

```python
# 일별x시간별 -> 일별로 병합
train1['ds'] = pd.to_datetime(train1['DateTime']).dt.date
train1 = train1.groupby('ds').sum().reset_index()
train1
```

```python
# 일별x시간별 -> 일별로 병합
train2['ds'] = pd.to_datetime(train2['DateTime']).dt.date
train2 = train2.groupby('ds').sum().reset_index()
train2
```

```python
# train1 + train2 병합하여, 하나의 train set으로 만들어 사용
train = pd.concat([train1, train2])
train
```

```python
train.reset_index(inplace=True)
train.drop('index', axis=1, inplace=True)
train
```

### year_month 급증 이유 탐색
1. 단순히 대회가 많이 개최 or 마감되어서?
2. 대회 개회 or 마감수는 적지만, 참가자가 많은 or 상금이많은 or 특정 주제 대회가 개회되어서?

```python
# '0000-00'
def year_month(x):
    x = str(x)
    return x[:7]
```

```python
info_compet['period_start_year_month'] = info_compet['period_start'].apply(year_month)
info_compet['period_end_year_month'] = info_compet['period_end'].apply(year_month)
info_compet['merge_deadline_year_month'] = info_compet['merge_deadline'].apply(year_month)
info_compet.head(2)
```

```python
# 대회 시작/종료일, 팀 병합 마감일 year_month별 count
tmp1 = info_compet.groupby(['period_start_year_month'])[['cpt_id']].count().reset_index()
tmp2 = info_compet.groupby(['period_end_year_month'])[['cpt_id']].count().reset_index()
tmp3 = info_compet.groupby(['period_end_year_month'])[['cpt_id']].count().reset_index()
tmp1.columns = ['year_month','period_start']
tmp2.columns = ['year_month','period_end']
tmp3.columns = ['year_month','merge_deadline']
```

```python
tmp1['year_month'] = tmp1['year_month'].astype(str)
tmp2['year_month'] = tmp2['year_month'].astype(str)
tmp3['year_month'] = tmp3['year_month'].astype(str)
```

```python
compet_cnt_year_month = pd.merge(tmp1, tmp2, on='year_month', how='outer')
compet_cnt_year_month = pd.merge(compet_cnt_year_month, tmp3, on='year_month', how='outer').sort_values(by='year_month')
compet_cnt_year_month.reset_index(inplace=True)
compet_cnt_year_month.drop('index', axis=1, inplace=True)
```

```python
compet_cnt_year_month.style.background_gradient(cmap='Blues')     # .style.bar()
```

```python
# 대회 시작일 병합 준비
period_start = info_compet[['period_start']].copy()
period_start.columns = ['ds']
period_start['ds'] = pd.to_datetime(period_start['ds']).dt.date
period_start['period_start'] = 1
period_start
```

```python
# 중복 날짜 제거
period_start.drop_duplicates(inplace=True)
period_start.shape
```

```python
# 'ds' type 맞춰주기
period_start['ds'] = period_start['ds'].astype(str)
train['ds'] = train['ds'].astype(str)
```

```python
# train + period_start 데이터 병합
train = pd.merge(train, period_start, on='ds', how='left')
train
```

```python
train.loc[train['period_start'].isnull(), 'period_start'] = 0
train['period_start'] = train['period_start'].astype(int)
train['period_start'].value_counts()
```

```python
# 대회 종료일 병합 준비
period_end = info_compet[['period_end']].copy()
period_end.columns = ['ds']
period_end['ds'] = pd.to_datetime(period_end['ds']).dt.date
period_end['period_end'] = 1
period_end
```

```python
# 중복 날짜 제거
period_end.drop_duplicates(inplace=True)
period_end.shape
```

```python
# 'ds' type 맞춰주기
period_end['ds'] = period_end['ds'].astype(str)
train['ds'] = train['ds'].astype(str)
```

```python
# train + period_end 데이터 병합
train = pd.merge(train, period_end, on='ds', how='left')
train
```

```python
train.loc[train['period_end'].isnull(), 'period_end'] = 0
train['period_end'] = train['period_end'].astype(int)
train['period_end'].value_counts()
```

```python
# 팀 병합 마감일 병합 준비
merge_deadline = info_compet[['merge_deadline']].copy()
merge_deadline.columns = ['ds']
merge_deadline['ds'] = pd.to_datetime(merge_deadline['ds']).dt.date
merge_deadline['merge_deadline'] = 1
merge_deadline
```

```python
# 결측치 제거
merge_deadline.drop(merge_deadline[merge_deadline['ds'].isnull()].index, inplace=True)
```

```python
# 중복 날짜 제거
merge_deadline.drop_duplicates(inplace=True)
merge_deadline.shape
```

```python
# 'ds' type 맞춰주기
merge_deadline['ds'] = merge_deadline['ds'].astype(str)
train['ds'] = train['ds'].astype(str)
```

```python
# train + deadline 데이터 병합
train = pd.merge(train, merge_deadline, on='ds', how='left')
train
```

```python
train.loc[train['merge_deadline'].isnull(), 'merge_deadline'] = 0
train['merge_deadline'] = train['merge_deadline'].astype(int)
train['merge_deadline'].value_counts()
```
