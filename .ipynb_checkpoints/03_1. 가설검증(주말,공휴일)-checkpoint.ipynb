# 대회 시작/종료/팀병합마감일의 높은/낮은 수치 경향성 차이 이유는?
**가설**
1. 주말, 공휴일의 영향으로 낮은 수치가 나타나는 것일까?
2. 대회 특성에 따라 낮게/높게 나타나는 추이가 나뉘지 않을까?      


(단, 가설 1,2에 해당되는 날이라도, 해당일에 대회가 2번이상 개최되었다면, 낮은 경향성으로 구분할 때 제외)

## 가설 1) 주말, 공휴일의 영향으로 낮은 수치가 나타나는 것일까?

### 주말

#### period_start

```python
# 같은날 대회 2개이상인 것도 포함
train2[(train2['week'] >= 5) & (train2['period_start'] >= 1)]
```

```python
tmp = train2[(train2['week'] >= 5) & (train2['period_start'] >= 1)][['ds','사용자','세션','신규방문자','페이지뷰']].copy()

# 추이 & event_point(period_start)
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=('사용자','세션','신규방문자','페이지뷰'), horizontal_spacing=0.08, vertical_spacing=0.12
)
fig.add_trace(go.Scatter(x=train2['ds'], y=train2['사용자'],name="사용자",showlegend=True, line=dict(color='#BDBDBD')), row=1, col=1)
fig.add_trace(go.Scatter(x=tmp['ds'], y = tmp['사용자'], mode = 'markers', name = '대회 시작일', marker=dict(size=5)), row=1, col=1)

fig.add_trace(go.Scatter(x=train2['ds'], y=train2['세션'],name="세션",showlegend=True, line=dict(color='#f08069')), row=1, col=2)
fig.add_trace(go.Scatter(x=tmp['ds'], y = tmp['세션'], mode = 'markers', name = '대회 시작일', marker=dict(size=5)), row=1, col=2)

fig.add_trace(go.Scatter(x=train2['ds'], y=train2['신규방문자'],name="신규방문자",showlegend=True, line=dict(color='#edab4e')), row=2, col=1)
fig.add_trace(go.Scatter(x=tmp['ds'], y = tmp['신규방문자'], mode = 'markers', name = '대회 시작일', marker=dict(size=5)), row=2, col=1)

fig.add_trace(go.Scatter(x=train2['ds'], y=train2['페이지뷰'],name="페이지뷰",showlegend=True, line=dict(color='#89d69e')), row=2, col=2)
fig.add_trace(go.Scatter(x=tmp['ds'], y = tmp['페이지뷰'], mode = 'markers', name = '대회 시작일', marker=dict(size=5, color='#f08069')), row=2, col=2)

fig.update_layout(height=600, width =1100, template='plotly_white', title_text='대회 시작일')  # template='ggplot2'
fig.show()
```

- 불규칙해서, 주말의 영향으로 낮은 수치가 나온다고 단정지을 수 없음
