**토픽모델링: k-means vs LDA**
- k-means도 토픽 모델링으로 이용 가능 (LDA와 같은 토픽 모델링과 k-means의 차이는 문서의 토픽 개수에 대한 가정)
    - **LDA**는 한 문서에 **여러 종류**의 토픽이 존재할 수 있다 가정 
        - 한 영화에 대한 여러 관점의 리뷰 -> 여러 토픽이 섞여있을 수 있음
    - **k-means**는 하나의 문서가 **하나의 토픽**이라 가정  
        - 주로 뉴스는 한 기사에 하나의 토픽이 있음 
- 따라서, 한 문서에 하나의 토픽이 존재한다고 가정한다면, 굳이 over-spec인 LDA를 이용할 필요는 없음    
- 출처: https://lovit.github.io/nlp/2018/09/27/pyldavis_kmeans/

## 데이터 불러오기 및 필요 컬럼만 남기기

```python
info_compet = pd.read_csv('data/info_compet.csv', encoding='cp949')
info_compet
```

```python
compet_start = info_compet[['period_start','prize','name','keyword']].copy()
compet_start
```

```python
# name, keyword 합친 후, 조건화 해서 대회 유형 분류
compet_start['document'] = compet_start['name'] + ' ' + compet_start['keyword']
```

## 대회 유형 분류

```python
# 수정된 조건 - 분류, 회귀, 시각화, 이미지/영상/음성 식별, 이상탐지, 문서 요약, 공정 최적화 
compet_start['type'] = np.nan

for i in compet_start.index:
    # 시각화 
    if '시각화' in compet_start['document'][i]:
        if ('분류' not in compet_start['document'][i]) and ('예측' not in compet_start['document'][i]):
            compet_start.loc[i, 'type'] = '시각화'
        elif ('예측' in compet_start['document'][i]) or ('회귀' in compet_start['document'][i]):
            compet_start.loc[i, 'type'] = '회귀'
        elif '분류' in compet_start['document'][i]:
            compet_start.loc[i, 'type'] = '분류'
    # 이미지/영상/음성 식별
    elif ('이미지' in compet_start['document'][i]) or ('영상' in compet_start['document'][i]) or ('음성' in compet_start['document'][i]):
        compet_start.loc[i, 'type'] = '이미지/영상/음성 식별'
    # 문서 요약
    elif '문서' in compet_start['document'][i]:
        compet_start.loc[i, 'type'] = '문서 요약'
    # 이상탐지
    elif '탐지 알고리즘' in compet_start['document'][i]:
        compet_start.loc[i, 'type'] = '이상 탐지'
    # 공정 최적화
    elif '공정' in compet_start['document'][i]:
        compet_start.loc[i, 'type'] = '공정 최적화'
    # 회귀
    elif ('예측' in compet_start['document'][i]) or ('회귀' in compet_start['document'][i]) or ('정형' in compet_start['document'][i]):
        if ('분류' not in compet_start['document'][i]):
            compet_start.loc[i, 'type'] = '회귀'
        else:
            compet_start.loc[i, 'type'] = '분류'
    elif ('MAE' in compet_start['document'][i]) or ('MSE' in compet_start['document'][i]):
        compet_start.loc[i, 'type'] = '회귀'
    # 분류
    elif '분류' in compet_start['document'][i]:
        compet_start.loc[i, 'type'] = '분류'
```

```python
# okt
okt = Okt()

stop_words_en = ['logloss','auc','rmse', 'mse','mae', 'map','rouge-n','accuracy','similarity','score', 'rouge',
                 'ai', 'dacon','loss','mspe','smape','tableau','powerbi','python','r','power','bi','spotfire','my','small'
                'gap','f','rmsle','smp','wrmsse','chrono','kl','divergence','tapr','csi','pinball']
stop_words_ko = ['빅데이터', '빅', '데이터','경진','대회','참가','중복', '불가', '제출', '인공','지능','시계열', '유저','산출',
                 '평가','심사','심사평','가능','월간','데이콘','데','이콘','회귀','시각화','시각','투표','대학','재학생','학생',
                 '정형','분류', '예측', '분석', '예측 분석', '이용','활용','로','를','및','공공','한국','제주','연구원','용량']

def text_preprocessing(documents):
    # 영문, 숫자 처리: 소문자로 변환 -> 토큰화
    doc_en = documents.lower()
    tokenized_doc_en = nltk.regexp_tokenize(doc_en, '[A-Za-z]+')
    # stop words 제거
    tokenized_doc_en = [word for word in tokenized_doc_en if word not in stop_words_en]
    
    # 한국어 처리: 명사만 추출
    tokenized_doc_ko = okt.nouns(documents) 
    # stop words 제거
    tokenized_doc_ko = [word for word in tokenized_doc_ko if word not in stop_words_ko]

    return tokenized_doc_en + tokenized_doc_ko
```

```python
# 클러스터 확인용 - 각 클러스터간의 핵심적인 단어 추출
def get_cluster_details(cluster_model, cluster_data, feature_names, cluster_num, top_n_features=10):
    cluster_details = {}
    center_feature_idx = cluster_model.cluster_centers_.argsort()[:,::-1]
    for cluster_num in range(cluster_num):
        cluster_details[cluster_num] = {}
        cluster_details[cluster_num]['cluster'] = cluster_num
        
        top_ftr_idx = center_feature_idx[cluster_num, :top_n_features]
        top_ftr = [feature_names[idx] for idx in top_ftr_idx]
        top_ftr_val = cluster_model.cluster_centers_[cluster_num, top_ftr_idx].tolist()
        
        cluster_details[cluster_num]['top_features'] = top_ftr
        cluster_details[cluster_num]['top_features_value'] = top_ftr_val
        
        filenames = cluster_data[cluster_data['cluster_label'] == cluster_num]['name']
        filenames = filenames.values.tolist()
        cluster_details[cluster_num]['filenames'] = filenames
    return cluster_details
```

```python
# 클러스터 확인용 - print 양식
def print_cluster_details(cluster_details):
    for cluster_num, cluster_detail in cluster_details.items():
        print(f'#####Cluster Num: {cluster_num}')
        print('상위 10개 feature단어들: \n', cluster_detail['top_features'])
        print(f"Cluster {cluster_num}으로 분류된 문서들: {cluster_detail['filenames'][:5]}")
        print('-'*20)
```

```python
# 여러개 k값 조정해가며 비교해보자!

inertia_list = []
k_list = range(2,17)
for k in k_list:
    model = KMeans(n_clusters=k, random_state=3)
    model.fit(X)
    inertia_list.append(model.inertia_)
```

```python
# elbow 기법
# 서로 다른 K개 클러스터 분류의 오차제곱값(SSE)를 확인할 수 있다. 변곡점이 되는 지점에서 K값을 결정하면 된다.
plt.figure(figsize=(15,6))
plt.plot(k_list, inertia_list)
plt.xlabel('그룹 수 (k)')
plt.ylabel('Inertia 값(SSE)')
plt.xticks(k_list)
plt.show()
```
