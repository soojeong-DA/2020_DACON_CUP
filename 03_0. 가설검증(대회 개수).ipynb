# 대회 시작/종료/팀병합마감일의 높은/낮은 수치 경향성 차이 이유는?
**가설**
1. 주말, 공휴일의 영향으로 낮은 수치가 나타나는 것일까?
2. 대회 특성에 따라 낮게/높게 나타나는 추이가 나뉘지 않을까?      


(단, 가설 1,2에 해당되는 날이라도, 해당일에 대회가 2번이상 개최되었다면, 낮은 경향성으로 구분할 때 제외)

## 가설 Base) 대회 2개 이상 개최/종료/팀병합마감된 날은 낮은 경향성 구분 제외   
- 대회 시작/종료/팀병함마감이 2개 이상인 날이 낮은 경향성을 보이지 않고, target 변수간의 양의 상관관계가 있다고 할 때
- 같은날 대회가 2번이상 개최되었다면, 가설 1,2에 해당되는 날이라도, 낮은 경향성으로 구분할 때 제외

##### 낮은 경향성 구분 시, 대회 2개 개최/종료/팀병합마감된 날은 제외 필요

### 대회 2개 이상 개최/종료/팀병합마감된 날 확인

#### 대회 시작일

```python
period_start_cnt = info_compet.groupby(['period_start'])[['cpt_id']].count().reset_index()
period_start_cnt.columns = ['ds','period_start_cnt']
period_start_cnt[period_start_cnt['period_start_cnt'] == 2]
```

```python
period_start_2_list = list(period_start_cnt[period_start_cnt['period_start_cnt'] == 2]['ds'])
period_start_2_list
```

```python
for d in train['ds']:
    if d in period_start_2_list:
        train.loc[train['ds'] == d, 'period_start'] = 2
```

```python
tmp = train[train['period_start'] == 1][['ds','사용자','세션','신규방문자','페이지뷰']].copy()
tmp2 = train[train['period_start'] == 2][['ds','사용자','세션','신규방문자','페이지뷰']].copy()

# 추이 & event_point
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=('사용자','세션','신규방문자','페이지뷰'), horizontal_spacing=0.08, vertical_spacing=0.12
)
fig.add_trace(go.Scatter(x=train['ds'], y=train['사용자'],name="사용자",showlegend=False, line=dict(color='#BDBDBD')), row=1, col=1)
fig.add_trace(go.Scatter(x=tmp['ds'], y = tmp['사용자'], mode = 'markers', name = '대회 시작일', marker=dict(size=5)), row=1, col=1)
fig.add_trace(go.Scatter(x=tmp2['ds'], y = tmp2['사용자'], mode = 'markers', name = '대회 시작일', marker=dict(size=5, color='#0091EA')), row=1, col=1)


fig.add_trace(go.Scatter(x=train['ds'], y=train['세션'],name="세션",showlegend=False, line=dict(color='#f08069')), row=1, col=2)
fig.add_trace(go.Scatter(x=tmp['ds'], y = tmp['세션'], mode = 'markers', name = '대회 시작일', marker=dict(size=5)), row=1, col=2)
fig.add_trace(go.Scatter(x=tmp2['ds'], y = tmp2['세션'], mode = 'markers', name = '대회 시작일', marker=dict(size=5, color='#9932CC')), row=1, col=2)

fig.add_trace(go.Scatter(x=train['ds'], y=train['신규방문자'],name="신규방문자",showlegend=False, line=dict(color='#edab4e')), row=2, col=1)
fig.add_trace(go.Scatter(x=tmp['ds'], y = tmp['신규방문자'], mode = 'markers', name = '대회 시작일', marker=dict(size=5)), row=2, col=1)
fig.add_trace(go.Scatter(x=tmp2['ds'], y = tmp2['신규방문자'], mode = 'markers', name = '대회 시작일', marker=dict(size=5, color='#9932CC')), row=2, col=1)

fig.add_trace(go.Scatter(x=train['ds'], y=train['페이지뷰'],name="페이지뷰",showlegend=False, line=dict(color='#89d69e')), row=2, col=2)
fig.add_trace(go.Scatter(x=tmp['ds'], y = tmp['페이지뷰'], mode = 'markers', name = '대회 시작일', marker=dict(size=5, color='#f08069')), row=2, col=2)
fig.add_trace(go.Scatter(x=tmp2['ds'], y = tmp2['페이지뷰'], mode = 'markers', name = '대회 시작일', marker=dict(size=5, color='#9932CC')), row=2, col=2)

fig.update_layout(height=600, width =1100, template='plotly_white', title_text='대회 시작일')
fig.show()
```

- 2개의 대회가 개최된 날에는 낮은 경향성을 보이지 않음
- 크게 급증하는 point가 모두 2개 대회 개최 특성을 가지고 있는 것은 아니지만, 2020-03-01에 대회 2개 개최되어 공휴일임에도 높은 수치 나타남

### 대회 개수와 traget변수간의 상관관계 확인

```python
plt.figure(figsize=(15, 7))

colormap = plt.cm.PuBu 
sns.heatmap(train[['사용자','세션','신규방문자','페이지뷰','period_start','period_end','merge_deadline']].corr(), 
            linewidths = 0.1, vmax = 1.0, square = True, cmap = colormap, linecolor = "white", annot = True, annot_kws = {"size" : 10.0})
plt.show()
```

- period_start, period_end는 사용자,세션,신규방문자,페이지뷰와 약한 양의 상관관계가 나타남
- merge_deadline은 4개의 traget 변수와 상관관계가 없는 것 같음 (period_start, period_end보다 상관성이 다소 약함)
